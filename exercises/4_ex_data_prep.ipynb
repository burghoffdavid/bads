```python
def make_odds_ratios(target, x):
    index_tuples = np.array(df.groupby(target)[x].value_counts().sort_index().index)
    print(index_tuples)
    index_names = []
    for t in index_tuples:
        index_names.append(t[1])
        
    value_counts = np.array(df.groupby(target)[x].value_counts().sort_index().tolist())
    value_counts = np.split(value_counts, 2)
    print(value_counts)
    ratios_df = pd.DataFrame(columns=index_names)
    ratios = []
    for idx, col in enumerate(ratios_df):
        ratios.append(round(value_counts[0][idx] / value_counts[1][idx], 2))
    
    ratios_df.loc[0] = ratios
    return ratios_df
```
